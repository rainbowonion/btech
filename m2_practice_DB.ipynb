```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('companies.csv')
```

```python
df.groupby('state').agg({'company_name': 'count'}).sort_values(by = 'company_name', ascending = False).reset_index()
fig, ax = plt.subplots(figsize=(8, 16))
sns.countplot(data=df, y='state')
plt.show()
```

```python
dfJobs = df.groupby('Industry').agg({'job_openings': 'sum'}).sort_values(by = 'job_openings', ascending = False).reset_index()
fig, ax = plt.subplots(figsize=(30, 5))
sns.barplot(data=dfJobs, x='Industry', y='job_openings')
plt.show()
```

```python
dfGrowth = df.groupby('founded').agg({'growth_percentage': 'mean'}).sort_values(by = 'growth_percentage', ascending = False).reset_index()
sns.lineplot(data=dfGrowth, x='founded', y='growth_percentage')
plt.show()
```

```python
sns.scatterplot(data=df, x='employees', y='estimated_revenues')
plt.show()
```

```python
dfPerYear = df.groupby('founded').agg({'company_name': 'count'}).sort_values(by = 'founded', ascending = True).reset_index()
sns.histplot(data=dfPerYear, x='founded')
plt.show()
```
