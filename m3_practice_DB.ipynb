```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup as bs
```

```python
response = rq.get('https://btech-data-analytics.github.io/bridgerland-technical-college/employees_3.html').text
```

```python
soup = bs(response, 'html.parser')
```

```python
soup.find('table').find_all('tr', class_='employee')[0].find_all('td')[1].text
```

```python
id = []
name = []
age = []

for employee in soup.find('table').find_all('tr', class_='employee'):
    id.append(employee.find_all('td')[0].text)
    name.append(employee.find_all('td')[1].text)
    age.append(employee.find_all('td')[2].text)

ageInt = list(map(int, age))

id, name, ageInt
```

```python
df = pd.DataFrame({
    'id':id,
    'name':name,
    'age':ageInt
})
df.head(10)
df.info()
```
