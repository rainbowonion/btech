```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup as bs
```

```python
response = rq.get('https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html').text
```

```python
soup = bs(response, 'html.parser')
```

```python
soup.find('table').find_all('tr', class_='book')[0].find_all('td')[0].text
```

```python
isbn = []
title = []
author = []
language = []
pages = []
topic = []
price = []

for book in soup.find('table').find_all('tr', class_='book'):
    isbn.append(book.find_all('td')[0].text)
    title.append(book.find_all('td')[1].text)
    author.append(book.find_all('td')[2].text)
    language.append(book.find_all('td')[3].text)
    pages.append(book.find_all('td')[4].text)
    topic.append(book.find_all('td')[5].text)
    price.append(book.find_all('td')[6].text)

pagesInt = list(map(int, pages))
price = [s.split('$')[1] for s in price]
priceFlt = list(map(float, price))

isbn, title, author, language, pagesInt, topic, priceFlt
```

```python
df = pd.DataFrame({
    'isbn':isbn,
    'title':title,
    'author':author,
    'language':language,
    'pages':pagesInt,
    'topic':topic,
    'price':priceFlt
    })
df.head(30)
df.info()
```

```python
df.groupby('author')['isbn'].count().sort_values(ascending=False).reset_index()
```

```python
df.groupby('topic')['isbn'].count().sort_values(ascending=False).reset_index()
```

```python
df.groupby('topic')['price'].mean().sort_values(ascending=False).reset_index()
```

```python
df.groupby('topic')['pages'].mean().sort_values(ascending=False).reset_index()
```
